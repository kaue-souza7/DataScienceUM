```python
import pandas as pd
```

```python
df = pd.read_csv('DataOfModule2/presidents.csv')
df.head()
```

```python
def splitname(row):

    row['First']=row['President'].split()[0]
    row['Last']=row['President'].split()[1]

    return row
```

```python
df = df.apply(splitname, axis='columns')
```

```python
df.head(

    
)
```

```python
del(df['First'])
del(df['Last'])
df.head()
```

```python
# pattern="(^[\w]*)(?:.* )([\w]*$)"
pattern="(^[\w]*)(?:. *)([\w]*$)"
df['President'].str.extract(pattern).head()
```

```python
pattern="(?P<First>^[\w]*)(?:. *)(?P<Last>[\w]*$)"


names = df['President'].str.extract(pattern).head()
```

```python
df['First'] = names['First']
df['Last'] = names['Last']
```

```python
df.head()
```

```python
# LIMPANDO A COLUNA "Born"
pattern='(^[\w]{3} [\w]{1,2}, [\w]{4})'
df['Born'] = df['Born'].str.extract(pattern)
df.head()


```

```python
pattern='(^[\w]{2}\s[\w]{5}, )([\w]{2,3})(\s[\w]{4})'
days = df['Age'].str.extract(pattern)
days.head()
```

```python
df.head()
```

```python
pattern='([\w]{2}\s[\w]{5},\s[\w]{2,3}\s[\w]{4})'
df['Age atstart of presidency'] = df['Age atstart of presidency'].str.extract(pattern)
df.head()
```

```python
# pattern='([\w]{2}\s[\w]{5},\s[\w]{2,3}\s[\w]{4}) ([\w]{3}\s[\w]{1,2},\s[\w]{4})'
pattern='([\w]{2}\s[\w]{5},\s[\w]{2,3}\s[\w]{4})([\w]{3}\s[\w]{1,2}.\s[\w]{4})'
date = df['Age atend of presidency'].str.extract(pattern)
del(date[0])
date.head()
```
