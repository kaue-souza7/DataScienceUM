# Tabelas Dinâmicas

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('DataOfModule3/cwurData.csv')
df.head()
```

```python
df[df['country'] == 'Brazil'].head()
```

```python
def applying_category(value):

    if value <= 100:
        value = 'First Tier Top University'
        return value

    elif value > 100 and value <= 200:
        value = 'Second Tier Top University'
        return value

    elif value > 200 and value <= 300:
        value = 'Third Tier Top University'
        return value
        
    value = 'Other University'
    return value

```

```python
df['Rank_level'] = df['world_rank'].apply(lambda x: applying_category(x))
```

```python
df.head()
```

```python
df = df.set_index(['country', 'Rank_level'])


df.groupby(level=(0, 1)).agg({'score':np.mean}).head(20)
```

```python
df.pivot_table(values='score', index='country', columns='Rank_level', aggfunc=[np.mean]).head()

# 1. df.pivot_table(...)
    # Aqui, df é um DataFrame do pandas, e você está chamando o método pivot_table para reorganizar 
    # os dados de df.

# 2. values='score'
    # Esta parte indica que você está interessado na coluna chamada 'score' para preencher os valores 
    # na nova tabela dinâmica.

# 3. index='country'
    # Os valores da coluna 'country' serão usados como índices (ou linhas) na nova tabela dinâmica. 
    # Ou seja, cada linha da tabela será rotulada com um país diferente.

# 4. columns='Rank_level'
    # Os valores da coluna 'Rank_level' serão usados como colunas na nova tabela dinâmica. Ou seja,
    # cada coluna da tabela será rotulada com um nível de ranking diferente.

# 5. aggfunc=[np.mean]
    # Esta parte define a função de agregação a ser aplicada para combinar os dados. Neste caso, você está
    # usando np.mean, que é a média. Isso significa que se houver múltiplos valores para a mesma combinação 
    # de 'country' e 'Rank_level', a média desses valores será calculada.
```

```python
df.pivot_table(values='score', index='country', columns='Rank_level', aggfunc=[np.mean, np.max]).head()
```

```python
new_df=df.pivot_table(values='score', index='country', columns='Rank_level', aggfunc=[np.mean, np.max],
               margins=True).head()
```
