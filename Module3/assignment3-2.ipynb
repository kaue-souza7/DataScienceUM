```python
import pandas as pd
import numpy as np
import re
import locale
```

```python
def answer_one():
    # YOUR CODE HERE
        #  //////////////////// ENERGY /////////////////////
    energia = pd.read_excel('DataOfModule3/Energy_Indicators.xls', skiprows=17)

    energia.drop(['Unnamed: 0', 'Unnamed: 1'], inplace=True, axis=1, errors='ignore')
    energia.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energia['Country'] = energia['Country'].replace({"Republic of Korea": "South Korea", 
                            "United States of America20": "United States", 
                            "United Kingdom of Great Britain and Northern Ireland19": "United Kingdom", 
                            "China, Hong Kong Special Administrative Region3n": "Hong Kong"
                        })
    energia['Energy Supply'] = energia['Energy Supply'] * 1_000_000
    
    
    def formating(item):
        item=str(item)
        item= re.sub(r'\(.*?\)', '', item)
        item= re.sub(r"\d+", '', item)
        item= re.sub(r'\s$', '', item)
        return item
    energia['Country'] = energia['Country'].apply(lambda x: formating(x))






        #  //////////////////// DGP /////////////////////

    dgp = pd.read_csv('DataOfModule3/world_bank.csv', skiprows=4)
    dgp['Country Name'] = dgp['Country Name'].replace({
                                                    "Korea, Rep.": "South Korea", 
                                                    "Iran, Islamic Rep.": "Iran",
                                                    "Hong Kong SAR, China": "Hong Kong"

                                                })
    dgp = dgp[['Country Name', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
                '2014', '2015']]
    dgp.columns = ['Country','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015']
    



        #  //////////////////// SCIMEM /////////////////////

    scimem = pd.read_excel('DataOfModule3/scimagojr-3.xlsx')
    scimem = scimem.head(16)


        #  //////////////////// CREAT DF /////////////////////

    df = pd.merge(scimem, energia ,how='inner' ,left_on='Country', right_on='Country')
    final_df = pd.merge(df,dgp, how='inner', left_on='Country', right_on='Country')
    final_df = final_df.set_index('Country')

    # df = pd.merge(scimem, energia, how='inner', on='Country')
    # df = pd.merge(df, dgp, how='inner', left_on='Country', right_on='Country Name')
    # # df.drop(columns='Country Name', inplace=True, errors='ignore')
    # # df.drop(columns='Indicator Code', inplace=True, errors='ignore')
    # df.set_index('Country', inplace=True)

    return final_df
    raise NotImplementedError()


answer_one()
```

```python
def answer_eleven():
    # YOUR CODE HERE
    df = answer_one()
    df = df.reset_index()
    df['Continent'] = df['Country']

    df['Continent'] = df['Country'].replace({'China':'Asia', 
                                            'United States':'North America', 
                                            'Japan':'Asia', 
                                            'United Kingdom':'Europe', 
                                            'Russian Federation':'Europe', 
                                            'Canada':'North America', 
                                            'Germany':'Europe', 
                                            'India':'Asia',
                                            'France':'Europe', 
                                            'South Korea':'Asia', 
                                            'Italy':'Europe', 
                                            'Spain':'Europe', 
                                            'Iran':'Asia',
                                            'Australia':'Australia', 
                                            'Brazil':'South America'
                                        })


    df['Population'] = df['Energy Supply'] / df['Energy Supply per Capita']


    # df.groupby('Continent').agg({'Population':(np.size, np.nanmean, np.nansum, np)})

    # OUUUU

    df = df.groupby('Continent')['Population'].agg(
                                                size='size',
                                                sum='sum',
                                                mean='mean',
                                                std='std'
                                            )

    return df

    raise NotImplementedError()
```

```python
def answer_twelve():
    # YOUR CODE HERE
    df = answer_one()
    df = df.reset_index()
    ContinetDict=({'China':'Asia', 
                                                'United States':'North America', 
                                                'Japan':'Asia', 
                                                'United Kingdom':'Europe', 
                                                'Russian Federation':'Europe', 
                                                'Canada':'North America', 
                                                'Germany':'Europe', 
                                                'India':'Asia',
                                                'France':'Europe', 
                                                'South Korea':'Asia', 
                                                'Italy':'Europe', 
                                                'Spain':'Europe', 
                                                'Iran':'Asia',
                                                'Australia':'Australia', 
                                                'Brazil':'South America'
                                            })
    df['Continent'] = [ContinetDict[c] for c in df['Country']]

    df['% Renewabl']=pd.cut(df['% Renewable'], 5)
    return df.groupby(['Continent','% Renewabl']).size()

    raise NotImplementedError()
```

# Question 13

```python
def answer_thirteen():
    # YOUR CODE HERE
    df = answer_one()
    df['Population'] = df['Energy Supply'] / df['Energy Supply per Capita']
    locale.setlocale(locale.LC_ALL, 'en_US.utf8')

    map_number = []
    for n in df['Population']:
        map_number.append(locale.format_string('%.2f', n, grouping=True))
    df['Population']=map_number

    return df['Population']

    raise NotImplementedError()
```
